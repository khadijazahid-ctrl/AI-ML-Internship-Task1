```python
# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to learn how to load, inspect, analyze, and visualize a dataset using Python. The Iris dataset is used to understand feature distributions, relationships between variables, and potential outliers through exploratory data analysis (EDA).

## Tools and Libraries
- Pandas
- Seaborn
- Matplotlib
```

```python
# Import required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Dataset Loading

The Iris dataset is loaded using Seaborn's built-in dataset collection.
```

```python
# Load Iris dataset

iris = sns.load_dataset('iris')

print("Dataset loaded successfully!")
```

```python
## Dataset Inspection

In this step, we inspect the dataset by checking:
- Shape of the dataset
- Column names
- First five rows
```

```python
# Shape of dataset

print("Shape of Dataset:")
print(iris.shape)

# Column names

print("\nColumn Names:")
print(iris.columns)

# First five rows

print("\nFirst Five Rows:")
print(iris.head())
```

```python
## Dataset Information

The .info() function provides:
- Number of rows and columns
- Data types
- Missing values
```

```python
print("Dataset Information:\n")
iris.info()
```

```python
## Summary Statistics

The .describe() function provides statistical information such as:
- Mean
- Standard Deviation
- Minimum Value
- Maximum Value
- Quartiles
```

```python
print("Summary Statistics:\n")
print(iris.describe())
```

```python
## Scatter Plot

This scatter plot helps visualize the relationship between sepal length and petal length for different flower species.
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='petal_length',
    hue='species'
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

```python
## Histograms

Histograms show the distribution of numerical features in the dataset.
```

```python
iris.hist(
    figsize=(10,8),
    bins=15
)

plt.suptitle("Feature Distributions")
plt.show()
```

```python
## Box Plots

Box plots help identify:
- Median values
- Spread of data
- Potential outliers
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=iris.iloc[:, :-1])

plt.title("Box Plot of Iris Features")

plt.show()
```

```python
## Results and Insights

### Observations

1. The Iris dataset contains 150 rows and 5 columns.
2. There are three species:
   - Setosa
   - Versicolor
   - Virginica
3. Petal measurements provide good separation between species.
4. Histograms show the distribution of each numerical feature.
5. Box plots reveal the spread of data and possible outliers.
6. The scatter plot demonstrates relationships between sepal and petal measurements.

### Conclusion

The Iris dataset is clean and well-structured for exploratory data analysis. Visualization techniques such as scatter plots, histograms, and box plots help uncover patterns and relationships among features.
```
